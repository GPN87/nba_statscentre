```python
from requests import get, auth
from pprint import pprint
import pandas as pd
from config import clientid, clientsecret
```

```python
baseurl = f'https://api.seatgeek.com/2/events?'

teams1 = ["boston-celtics",
"brooklyn-nets",
"new-york-knicks",
"philadelphia-76ers",
"toronto-raptors",
"golden-state-warriors",
"los-angeles-clippers",
"los-angeles-lakers",
"phoenix-suns",
"sacramento-kings",
"chicago-bulls",
"cleveland-cavaliers",
"detroit-pistons",
"indiana-pacers",
"milwaukee-bucks",
"dallas-mavericks",
"houston-rockets",
"memphis-grizzlies",
"san-antonio-spurs",
"atlanta-hawks",
"charlotte-hornets",
"miami-heat",
"orlando-magic",
"washington-wizards",
"denver-nuggets",
"minnesota-timberwolves",
"oklahoma-city-thunder",
"portland-trail-blazers",
"utah-jazz",
"new-orleans-pelicans"]


tix_prices = {}
home_coords = {}
next_matches = {}


for team in teams1:
    tix_prices[team]=[]
    home_coords[team]=[]
    next_matches[team]=[]
    params = {
        "client_id": clientid,
        "client_secret": clientsecret,
        "performers.slug":team
    }
    
    response = get(url=baseurl,params=params)
    data = response.json()
    events = data['events']
    for event in events:
        
        if event['performers'][0]['slug'] == team:
            avg_price = event['stats']['average_price']
            location = event['performers'][0]['location']
            
            tix_prices[team].append(avg_price)
            pprint(event['short_title'])
            home_coords[team] = location
            
        if event['performers'][0]['slug'] == team: 
            nextmatch = event['short_title']
            next_matches[team]=nextmatch
            break
```

```python
next_matches
```

```python
nextgame_df=pd.DataFrame(list(next_matches.items()),columns=['team_name','next_homegame'])
nextgame_df['team_name']=nextgame_df['team_name'].str.replace('-'," ").str.title()
```

```python
nextgame_df
```

```python
home_coords_list = {}
for key,value in home_coords.items():
    new_value = [value['lat'],value['lon']]
    home_coords_list[key] = new_value
```

```python
coord_df = pd.DataFrame.from_dict(home_coords_list, orient='index', columns=['lat', 'lon'])
coord_df.index.name='team_name'
coord_df.reset_index(inplace=True)

coord_df['team_name']=coord_df['team_name'].str.replace('-'," ").str.title()
```

```python
prices_df=pd.DataFrame(list(next_homeprice.items()),columns=['team_name','avg_price_next_homegame'])
prices_df['team_name']=prices_df['team_name'].str.replace('-'," ").str.title()
prices_df
```

```python
mergeI = coord_df.merge(nextgame_df,on="team_name").merge(prices_df,on="team_name")
mergeI['team_name'] = mergeI['team_name'].replace('Philadelphia 76Ers', 'Philadelphia 76ers')
```
